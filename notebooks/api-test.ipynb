```python
from datetime import datetime
from time import sleep
import hashlib
import json


for i in range(10):
    print(f'Making request {i} at {datetime.now()}')
    data = fetch_data()
    if data:
        result = data.get("d", {}).get("result", {})
        if not result:
            print('No result in response')
            continue

        created_at = result.get("@CreationTime", None)
        print(created_at)
        trains = result.get("Trains", {}).get("Train", [])
        trains_json = json.dumps(trains)
        trains_hash = hashlib.sha256(trains_json.encode('utf8')).hexdigest()
        print(trains_hash)
    sleep(1)
    print()
```
